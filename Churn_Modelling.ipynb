# Importing all libraries

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# Load our  dataset

```python
#loading churn modeling dataset
data = pd.read_csv('Churn_Modelling.csv')
data
```

# Data cleaning and data preprocessing

```python
#removing unnecessary features
data = data.drop(['RowNumber','CustomerId','Surname'], axis=1)
```

```python
data.isnull().sum()
```

```python
#Data processing for catagorical features Geography and gender
geo = pd.get_dummies(data['Geography'],drop_first=True)
gen = pd.get_dummies(data['Gender'],drop_first=True)
```

```python
data = pd.concat([data, geo, gen], axis=1)
data = data.drop(['Geography','Gender'], axis=1)
```

```python
y = data['Exited']
x = data.drop(['Exited'], axis=1)
```

# Split data into training and testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Creating ANN Model

```python
model = Sequential()

#input layer
model.add(Dense(units=14, kernel_initializer='he_uniform', activation='relu', input_dim=14))

#hidden layer 1
model.add(Dense(50, kernel_initializer='he_uniform', activation='relu'))

#dropout layer
model.add(Dropout(0.2))

#output layer
model.add(Dense(1, kernel_initializer='glorot_uniform', activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size=10, epochs=300)
```

# Evaluating model

```python
#evaluate model and checking for accuracy
model.evaluate(x_test, y_test)
```
